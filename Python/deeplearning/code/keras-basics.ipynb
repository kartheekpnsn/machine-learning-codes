## Keras with Tensorflow

- Tutorial from [here](http://cv-tricks.com/tensorflow-tutorial/keras/)
- Keras is a high-level python API which can be used to quickly build and train neural networks using either Tensorflow or Theano as back-end.
- Keras is being hailed as the future of building neural networks. Here are some of the reasons for its popularity:
  - **Light-weight and quick**: Keras is designed to remove boilerplate code. Few lines of keras code will achieve so much more than native Tensorflow code. You can easily design both CNN and RNNs and can run them on either GPU or CPU. 
  - **Emerging possible winner**: Keras is an API which runs on top of a back-end. This back-end could be either Tensorflow or Theano. Microsoft is also working to provide CNTK as a back-end to Keras.
- Keras has two distinct ways of building models:
  1. **Sequential models**: This is used to implement simple models. You simply keep adding layers to the existing model. 
  2. **Functional API**: Keras functional API is very powerful and you can build more complex models using it, models with multiple output, directed acyclic graph etc.

### 1) Keras - Sequential Models

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Conv2D, MaxPooling2D, Flatten, Dropout
model = Sequential()
```

#### 1.1) Steps
1. The first layer of a network reads the training data. So, we need to specify the size of images/training data that we are feeding the network. So, the input_shape parameter is used to specify the shape of input data: (the input layer is a convolutional layer which takes input images of 224 x 224 x 3.)
 - `model.add(Conv2D(32, (3, 3), activation = 'relu', input_shape = (224, 224, 3)))`
2. Convolutional layer: Here, we shall add a layer with 64 filters of size 3*3 and use relu activations after that.
 - `model.add(Conv2D(64, (3, 3), activation = 'relu')`
3. Maxpooling layer: Specify the type of layer and specify the pool size and you are done. How cool is that!
 - `model.add(MaxPooling2D(pool_size = l_size = (2, 2)))`
4. FullyConnected layer:  It’s called Dense in Keras. Just specify the number of outputs and you are done.
 - `model.add(Dense(256, activation = 'relu'))`
5. Dropout: 
 - `model.add(Dropout(0.5))`
6. Flatten layer:
 - `model.add(Flatten())`
7. Once you have specified the architecture of the network, you need to specify the method for back-propagation by choosing an optimizer( like rmsprop or adagrad) and specify the loss(like categorical_crossentropy ). We use compile function to do that in Keras.
 - `model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop')`
 - **[OR]**
 - `from keras.optimizers import SGD`
 - `sgd = SGD(lr = 0.01, decay = 1e-6, momentum = 0.9, nesterov = True)`
 - `model.compile(loss = 'categorical_crossentropy', optimizer = sgd)`
8. Now that we have created the model; let’s feed the data to the model via the fit function. You can also specify the batch_size and the maximum number of epochs you want training to go on.
 - `model.fit(x_train, y_train, batch_size = 32, epochs = 10,validation_data = (x_val, y_val))`
9. Evaluation:
 - `score = model.evaluate(x_test, y_test, batch_size=32)`

#### 1.2) Solve a linear regression problem with an example

```python
# # create training data
# x lies between -1 and 1
# y = 3 * x + noise
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy as np

trX = np.linspace(-1, 1, 101)
trY = 3 * trX + np.random.randn(*trX.shape) * 0.33
```

```python
# # create a model
model = Sequential()
model.add(Dense(input_dim = 1, units = 1, kernel_initializer = 'uniform', activation = 'linear')) 
# where:
# input_dim = number of independent features
# units = output_dim
# activation = linear
```

```python
weights = model.layers[0].get_weights()
w_init = weights[0][0][0]
b_init = weights[1][0]
print('Linear regression model is initialized with weights w: %.2f, b: %.2f' % (w_init, b_init))
```

```python
model.compile(optimizer='sgd', loss='mse') # uses stochastic gradient descent and loss = mse
```

```python
model.fit(trX, trY, epochs = 200, verbose = 1)
```

```python
weights = model.layers[0].get_weights()
w_final = weights[0][0][0]
b_final = weights[1][0]
print('Linear regression model is trained to have weight w: %.2f, b: %.2f' % (w_final, b_final))
```

#### 1.3) Solve a multiple linear regression problem with an example

```python
trX1 = np.linspace(-1, 1, 101)
trX2 = np.linspace(0, 1, 101)
trX = np.column_stack((trX1, trX2))
trY = 3 * trX1 + 2 * trX2 + 0.33
```

```python
# # create a model
model = Sequential()
model.add(Dense(5, input_dim = 2))
model.add(Activation('relu'))
model.add(Dense(1))
# model.add(Activation('sigmoid'))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
model.fit(trX, trY, epochs=200, batch_size=10)
```

```python
model.evaluate(trX, trY)
```

```python
print(model.predict(trX[:5]))
print(trY[:5])
```

### 2) Keras - Functional API
- For solving real world complex problems
- Functional API allows you to call models like a function as we do for a layer. So, you can build a small model which can be repeated multiple times to build the complete network with even fewer lines of code.
- By calling a model on a tensor, you can reuse the model as well as weights. 
- In most of the popular neural networks, we have a mini-network(a few layers arranged in a certain way like Inception module in GoogLeNet, fire module in squeezeNet) and this mini-network is repeated multiple times
- Required Imports
 - `from keras.models import Model`
 - `from keras.layers import Input`
- Now, you start by specifying the input, as opposed to mentioning the input at the end of the fit function, as done in Sequential models.
 - `digit_input = Input(shape=(28, 28,1))`
- We apply a convolutional layer using the Functional API, we shall have to specify the variable on which we want to apply the layer.
 - `x = Conv2D(64, (3, 3))(digit_input)`
 - `x = Conv2D(64, (3, 3))(x)`
 - `x = MaxPooling2D((2, 2))(x)`
 - `out = Flatten()(x)`
- Finally, we create a model by specifying the input and output.
 - `vision_model = Model(digit_input, out)`

#### VGG16 Network

- VGG convolutional neural network was proposed by a research group at Oxford in 2014. 
- This network was once very popular due to its simplicity and some nice properties like it worked well on both image classification as well as detection tasks.
- VGG achieved 92.3% top-5 accuracy in ILSVRC 2014 but was not the winner.
- It has a few variants, one of the most popular ones is vgg-16 which has 16 layers.
- As you can see that it takes an input of 224 x 224 x 3.

![alt text](http://cv-tricks.com/wp-content/uploads/2017/05/vgg16.png "VGG16 Architecture")

```python
# # just to define architecture
img_input = Input(shape=(224, 224, 3))
# Block 1
x = Conv2D(64, (3, 3), activation='relu', padding='same', name='block1_conv1')(img_input)
x = Conv2D(64, (3, 3), activation='relu', padding='same', name='block1_conv2')(x)
x = MaxPooling2D((2, 2), strides=(2, 2), name='block1_pool')(x)

# Block 2
x = Conv2D(128, (3, 3), activation='relu', padding='same', name='block2_conv1')(x)
x = Conv2D(128, (3, 3), activation='relu', padding='same', name='block2_conv2')(x)
x = MaxPooling2D((2, 2), strides=(2, 2), name='block2_pool')(x)

# Block 3
x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv1')(x)
x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv2')(x)
x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv3')(x)
x = MaxPooling2D((2, 2), strides=(2, 2), name='block3_pool')(x)

# Block 4
x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block4_conv1')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block4_conv2')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block4_conv3')(x)
x = MaxPooling2D((2, 2), strides=(2, 2), name='block4_pool')(x)

# Block 5
x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block5_conv1')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block5_conv2')(x)
x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block5_conv3')(x)
x = MaxPooling2D((2, 2), strides=(2, 2), name='block5_pool')(x)

x = Flatten(name='flatten')(x)
x = Dense(4096, activation='relu', name='fc1')(x)
x = Dense(4096, activation='relu', name='fc2')(x)
x = Dense(3, activation='softmax', name='predictions')(x) # 3 = number of classes
```

```python
# use existing architecture
import numpy as np
from keras import applications
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
from keras.preprocessing import image
```

```python
# build the VGG16 network
model = applications.VGG16(weights='imagenet')
img = image.load_img('..//data//BFL103B-L.jpg', target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
preds = model.predict(x)
all_results = decode_predictions(preds)
for results in all_results: 
    for result in results:
        print('\nProbability %0.2f%% => [%s]' % (100*result[2], result[1]))
```

### ADAM Optimizer standard values
- TensorFlow: learning_rate=0.001, beta1=0.9, beta2=0.999, epsilon=1e-08.
- Keras: lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0.
